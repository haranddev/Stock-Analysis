```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
from sklearn.preprocessing import StandardScaler
```

```python
train_set=pd.read_csv('JPM1.csv')
test_set=pd.read_csv('JPM_test.csv')
```

```python
scaler = StandardScaler()
scaler.fit(train_set)
```

```python
scaled_data = scaler.transform(train_set)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=train_set['MACD'],cmap='plasma')
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component')
```

```python
pca.components_
```

```python
df_comp = pd.DataFrame(pca.components_,columns=train_set.columns)
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df_comp,cmap='plasma',)
```
